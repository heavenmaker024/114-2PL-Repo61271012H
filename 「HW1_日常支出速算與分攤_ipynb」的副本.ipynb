#日常支出速算與分攤（作業一）
- 目標：從 Sheet 讀「消費紀錄」→ 計總額/分類小計/AA 分攤 → 寫回 Sheet Summary 分頁。
- AI 點子（可選）：請模型總結本週花錢習慣與建議（例如「外食過多」）。
- Sheet 欄位：date, category, item, amount, payer

GoogleSheet: https://docs.google.com/spreadsheets/d/1pn9swy4jQIXOVdrNNigq7MeR_PlaC5NxKWeT4q47LW8/edit?gid=0#gid=0

```python
# -*- coding: utf-8 -*-
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd
import datetime

# ==========================================
# 1. 驗證 Google 帳號授權
# ==========================================
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# ==========================================
# 2. 連結到指定的 Google 試算表
# ==========================================
# 填入你提供的新 Google Sheet 網址
SHEET_URL = 'https://docs.google.com/spreadsheets/d/1pn9swy4jQIXOVdrNNigq7MeR_PlaC5NxKWeT4q47LW8/edit?gid=0#gid=0'
gsheets = gc.open_by_url(SHEET_URL)

# 取得指定的分頁 (預設為 '工作表1')
worksheet = gsheets.worksheet('工作表1')

# 將現有資料載入 DataFrame 方便查看
sheets_data = worksheet.get_all_values()
if len(sheets_data) > 0:
    df = pd.DataFrame(sheets_data[1:], columns=sheets_data[0])
    print("目前試算表前 5 筆資料：")
    print(df.head())
else:
    print("這是一個空白的試算表。")

# ==========================================
# 3. 讓使用者輸入新資料
# ==========================================
print("\n--- 開始新增消費紀錄 ---")
date_str = input("請輸入日期 (格式：YYYY-MM-DD): ")
# 檢查日期格式是否正確
datetime.datetime.strptime(date_str, '%Y-%m-%d')

time_str = input("請輸入時間 (格式：HH:MM): ")
# 檢查時間格式是否正確
datetime.datetime.strptime(time_str, '%H:%M')

category = input("請輸入分類： ")
item = input("請輸入品項: ")
amount = float(input("請輸入金額: "))

# 為了配合你的 9 個欄位，加入以下選填欄位
payer = input("請輸入付款人 (可留白直接按 Enter): ")
location = input("請輸入地點 (可留白直接按 Enter): ")
pay_method = input("請輸入支付方式 (可留白直接按 Enter): ")
note = input("請輸入備註 (可留白直接按 Enter): ")

# ==========================================
# 4. 將新資料寫入 Google 試算表
# ==========================================
# 將新資料打包成一個 List，嚴格對應試算表的 9 個欄位
# 順序：[日期, 時間, 分類, 品項, 金額, 付款人, 地點, 支付方式, 備註]
new_row_data = [date_str, time_str, category, item, amount, payer, location, pay_method, note]

# 使用 append_row 將這「一筆」新資料加到最底下
worksheet.append_row(values=new_row_data, value_input_option='USER_ENTERED')

print(f"\n✅ 資料已成功寫入 Google 工作表！")
print(f"新增的內容為：{new_row_data}")
```
